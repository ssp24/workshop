```python
import pandas as pd
import spacy
from spacy import displacy
import os
import string
import nltk
from nltk.corpus import stopwords
from collections import Counter
```

```python
df = pd.read_csv("otra_alemania_content.csv", encoding="UTF-8")
df
```

```python
nltk.download('stopwords')

# Load German stopwords
stopger = stopwords.words('german')

# Additional stopwords
new_stopwords = [
        'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', '«', '»',
        'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z', 'vgl', '\x97', '•', '■', 'v',
        'beim', 'de','—','ge','la','be','en','que','el','ten','ver','gen','sei','nen','del','nen', 'se','schen','un','land','te','ei','aires'
    ]
stopger.extend(new_stopwords)

# Define a function to clean a single text entry
def clean_text(entry):
    entry = str(entry)
    # Remove punctuation
    entry = entry.translate(str.maketrans('', '', string.punctuation))
    # Tokenize and remove numbers
    words = [word for word in entry.split() if not word.isdigit()]
    # Remove stopwords
    cleaned_text = [word for word in words if word.lower() not in stopger]
    return ' '.join(cleaned_text)
```

```python
df['cleaned_text'] = df['text'].apply(clean_text)
df
```

### Lemmatization with spacy: 

```python
def get_lemma(doc):
    return [(token.lemma_) for token in doc]
```

### Part of speech tagging: 

```python
def get_pos(doc):
    return [(token.pos_, token.tag_) for token in doc]
```

```python
df['POS'] = df['token'].apply(get_pos)
df
```

(PROPN, NE) = "Poper noun, that has been recognized as a named entity!"

PROPN: This is a part-of-speech tag indicating that the word is a proper noun. Proper nouns are specific names of people, places, organizations, etc., and they are typically capitalized in English and many other languages.

NE: This stands for Named Entity. Named Entity Recognition (NER) is a process in natural language processing (NLP) where entities such as names of people, organizations, locations, dates, etc., are identified and classified in text.

## Get all the words that have been recognized as (PROPN, NE):

```python
def extract_propn_ne(df):
    propn_ne_words = []  # List to store extracted words

    for tokens, pos_tags in zip(df['token'], df['POS']):
        for token, pos in zip(tokens, pos_tags):
            if pos == ("PROPN", "NE"):
                propn_ne_words.append(token)
    
    return propn_ne_words
```

```python
propn_ne_words = extract_propn_ne(df)
```

```python
print(len(propn_ne_words))
print(propn_ne_words)
```

```python
def create_ner_dataframe(tokens):
    # Prepare data for the DataFrame
    data = []
    for token in tokens:
        # Extract the text and NER label of the token
        word = token.text
        label = token.ent_type_  # Get the NER label
        data.append((word, label))
    
    # Create a DataFrame
    df = pd.DataFrame(data, columns=['Token', 'NER_Label'])
    
    return df
```

```python
df_ner = create_ner_dataframe(propn_ne_words)
df_ner
```

## Topic Modeling:

Latent Dirichlet Allocation (LDA):¶

```python
import gensim
from gensim import corpora
from gensim.parsing.preprocessing import preprocess_string
```

```python
test_df = df.head(500)
test_df
```

```python
lemmatized_text = ' '.join(test_df['lemma'].explode())
print(lemmatized_text)
```

```python
#Erstellen eines Dictionary und einer Doc-Term-Matrix:
processed_text = preprocess_string(lemmatized_text)
dictionary = corpora.Dictionary([processed_text])
doc_term_matrix = [dictionary.doc2bow(doc.split()) for doc in processed_text]
```

```python
# Apply LDA for topic modeling
lda_model = gensim.models.LdaModel(doc_term_matrix, num_topics=4, id2word=dictionary, passes=15)
```

```python
# Visualize the topics
import pyLDAvis.gensim_models
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, doc_term_matrix, dictionary)
vis
```
